
### **1. Inheritance: Specialized Courses**
Inheritance allows courses to share common features while also having specific characteristics.

1. **Base Class: Course**
   - Common attributes: `course_name`, `credits`, `instructor`
   - Common methods: `get_syllabus()`, `enroll_student()`

2. **Derived Classes:**
   - **ProgrammingCourse** (inherits from Course):
     - Additional attribute: `programming_language`
     - Method: `get_practical_exercises()`
   
   - **MathematicsCourse** (inherits from Course):
     - Additional attribute: `math_field` (e.g., Calculus, Statistics)
     - Method: `get_formula_sheet()`

*Code Version with Inheritance*

```python
# Base Class: Course
class Course:
    def __init__(self):
        # initialize attribute
        self.course_name = ""
        self.credits = 0
        self.instructor = ""
        self.enrolled_students = []  # List to hold enrolled students
        self.syllabus=""
    # methods to get input from user
    def get_syllabus(self):
        self.syllabus=input("Enter Syllabus: ")

    def enroll_student(self):
        student_name = input("Enter student name : ")
        self.enrolled_students.append(student_name)

    def get_course_details(self):
        self.course_name = input("Enter Course Name: ")
        self.credits = int(input("Enter Course Credit: "))
        self.instructor = input("Enter Course Instructor: ")


     # method to display user input
    def display_course_details(self):
        print("\n***************************************************")
        print("Course Name:", self.course_name)
        print("Course Credit:", self.credits)
        print("Course Instructor:", self.instructor)


#----------------------------------------------------Same Parent Class----------
#Create child/derived Class
class ProgrammingCourse(Course):
    def __init__(self):
        super().__init__()  #inherit previous attributes from Parent Class
       #add new attributes
        self.programming_language = ""
        self.exercises = ""

#add new method
    def get_practical_exercises(self):
        self.exercises=input("Enter Project Excersice: ")

    def get_course_details(self):
        print("---------Enter details of Programming Course---------\n")
        super().get_course_details() #get course, credit, instructor from Parent
        super().enroll_student() #get student from Parent
        self.programming_language = input("Enter Programming Language: ")
        super().get_syllabus()


    def display_course_details(self):
        print("\n***************************************************")
        print("-------Programming Course Details-------")
        super().display_course_details() #print course, credit, instructor from Parent
        print("Enrolled Students:", ", ".join(self.enrolled_students))  # Show all students
        print("Programming Language:", self.programming_language)
        print("Syllabus:", self.syllabus) #print syllabus from Parent Class
        print("Practical exercises:", self.exercises )


#create another child/derived calss
class MathematicsCourse(Course):
    def __init__(self):
        super().__init__()
        self.math_field = ""    #add new attributes
        self.formula_sheet = ""

#add new method
    def get_formula_sheet(self):
       self.formula_sheet=input("Enter Formula Sheet: ")

    def get_course_details(self):
       print("---------Enter details of Mathematics Course---------\n")
       super().get_course_details()#get course, credit, instructor from parent
       super().enroll_student()    #get student from parent
       super().get_syllabus()  #get syllabus from parent
       self.math_field = input("Enter Math Field: ")

    def display_course_details(self):
       print("\n***************************************************")
       print("-------Mathematics Course Details-------")
       super().display_course_details()#print course, credit, instructor from parent
       print("Enrolled Students:", ", ".join(self.enrolled_students))  # Show all students
       print("Syllabus:", self.syllabus) #print syllabus from Parent Class
       print("Math Field:", self.math_field)
       print("Formula Sheet:", self.formula_sheet)

# Create Programming course object course
oop = ProgrammingCourse()  # Initialize with empty values
oop.get_course_details()
oop.get_practical_exercises()
oop.display_course_details() #diplay Output

# Create Mathematics course object course
calculus = MathematicsCourse()  # Initialize with empty values
calculus.get_course_details()
calculus.get_formula_sheet()
calculus.display_course_details() #display output
```


### **3. Encapsulation:**
- **Private Attributes**: These are the details of a class (like course information) that you dont want everyone to change directly. You hide them by using a double underscore (`__`).
  - **Example**: `__course_name`, `__credits`, and `__instructor` in our **Course** class.

- **Public Methods**: These are the functions that allow you to access or change the private attributes safely. Think of them as the "safe doors" to get in and out of the private data.
  - **Example**: Methods like `get_course_name()`, `get_credits()`, and `enroll_student()` let you interact with the course details without exposing them directly.  Or
  - **Example**: Common Methods like `get_course_details()`, `display_course_details()` let you interact with the course details without exposing them directly.

### **4.Abstraction: Hiding Complexity**
Abstraction allows you to interact with a course without needing to understand all the underlying details.

- You can enroll in a course by simply calling the `enroll_student()` method without needing to know how the enrollment process works behind the scenes (like database operations, checks, etc.).<br>

*Code Version of Inheritance, Polymorphism, Encapsulation & Abstraction*

```python
# Base Class: Course
from abc import ABC, abstractmethod
class Course(ABC): #abstract syntax
    def __init__(self):
        # initialize attribute
        self.__course_name = ""
        self.__credits = 0
        self.__instructor = ""
        self._enrolled_students = []  # List to hold enrolled students
        self._syllabus=""

    # methods to get input from user
    def get_syllabus(self):
        self._syllabus=input("Enter Syllabus: ")

    @abstractmethod
    def enroll_student(self):
        pass

    def get_course_details(self):
        self.__course_name = input("Enter Course Name: ")
        self.__credits = int(input("Enter Course Credit: "))
        self.__instructor = input("Enter Course Instructor: ")

    def get_schedule(self):
        pass

     # method to display user input
    def display_course_details(self):
        print("\n***************************************************")
        print("Course Name:", self.__course_name)
        print("Course Credit:", self.__credits)
        print("Course Instructor:", self.__instructor)


#----------------------------------------------------Same Parent Class----------
#Create child/derived Class
class ProgrammingCourse(Course):
    def __init__(self):
        super().__init__()  #inherit previous attributes from Parent Class
       #add attributes
        self.programming_language = ""
        self.exercises = ""
        self.coding_assignments = ""
        self.lab_sessions = ""


    # Implementing the abstract method from parent class
    def enroll_student(self):
        _student_name = input("Enter student name to enroll: ")
        self._enrolled_students.append(_student_name)
#add method
    def get_practical_exercises(self):
        self.exercises=input("Enter Project Excersice: ")

    def get_course_details(self):
        print("---------Enter details of Programming Course---------\n")
        super().get_course_details() #get course, credit, instructor from Parent
        self.enroll_student() #get student from abstract method
        self.programming_language = input("Enter Programming Language: ")
        super().get_syllabus()

#add method
    def get_schedule(self):
        self.coding_assignments = input("Enter Coding Assignments: ")
        self.lab_sessions = input("Enter Lab Sessions: ")

    def display_course_details(self):
        print("\n***************************************************")
        print("-------Programming Course Details-------")
        super().display_course_details() #print course, credit, instructor from Parent
        print("Enrolled Students:", ", ".join(self._enrolled_students))  # Show all students
        print("Programming Language:", self.programming_language)
        print("Syllabus:", self._syllabus) #print syllabus from Parent Class
        print("Practical exercises", self.exercises )
        print("Coding Assignments:", self.coding_assignments) #new attribute
        print("Lab Sessions:", self.lab_sessions)             #new attribute

#create another child/derived calss
class MathematicsCourse(Course):
    def __init__(self):
        super().__init__()
        self.math_field = ""
        self.formula_sheet = ""
        self.theory_lectures = ""          #add new attributes
        self.problem_solving_sessions = "" #add new attributes


# Implementing the abstract method from parent class
    def enroll_student(self):
        _student_name = input("Enter student name to enroll: ")
        self._enrolled_students.append(_student_name)
#add method
    def get_formula_sheet(self):
       self.formula_sheet=input("Enter Formula Sheet: ")

    def get_course_details(self):
       print("---------Enter details of Mathematics Course---------\n")
       super().get_course_details()#get course, credit, instructor from parent
       self.enroll_student()    #get student from parent
       super().get_syllabus()  #get syllabus from parent
       self.math_field = input("Enter Math Field: ")

#add method
    def get_schedule(self):
       self.theory_lectures=input("Enter Theory Lectures: ")
       self.problem_solving_sessions=input("Enter Problem Solving Session: ")

    def display_course_details(self):
       print("\n***************************************************")
       print("-------Mathematics Course Details-------")
       super().display_course_details()#print course, credit, instructor from parent
       print("Enrolled Students:", ", ".join(self._enrolled_students))  # Show all students
       print("Syllabus:", self._syllabus) #print syllabus from Parent Class
       print("Math Field:", self.math_field)
       print("Formula Sheet:", self.formula_sheet)
       print("Theory Lectures:", self.theory_lectures)
       print("Problem Solving Sessions:", self.problem_solving_sessions)

# Create Programming course object course
oop = ProgrammingCourse()  # Initialize with empty values
oop.get_course_details()
oop.get_practical_exercises()
oop.get_schedule()
oop.display_course_details()

# Create Mathematics course object course
calculus = MathematicsCourse()  # Initialize with empty values
calculus.get_course_details()
calculus.get_formula_sheet()
calculus.get_schedule()
calculus.display_course_details()
```
