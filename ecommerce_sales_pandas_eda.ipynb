# B) Exploratory Data Analysis (EDA)

## 1.0 Monthly Sales

To display Monthly sales, we need
- Month from orders DataFrame/Table
- Revenue from sales DataFrame/Table

To get data from two DataFrames, we merge both DataFrames

Display Columns of Both DataFrames

```python
sales.columns
```

```python
orders.columns
```

New Column, Month

We see that there is no Column of Month in orders DataFrame

So, we have to extract month from OrderDate and save into new Column Month  


```python
# Convert 'OrderDate' to datetime objects
orders['OrderDate'] = pd.to_datetime(orders['OrderDate'])

# Extract the month from the 'OrderDate' column
orders['Month'] = orders['OrderDate'].dt.month

# Print the updated dataframe to see the new 'Month' column
orders
```

Merge Sales and Orders DataFrames

```python
merge_sales_orders = pd.merge(sales, orders, left_on='OrderID', right_on='TransactionID')
merge_sales_orders
```

Monthly Sales by groupby()

Now, we can apply groupby() on merge data `merge_sales_orders` to get sales according to month

```python
monthly_sales=merge_sales_orders.groupby(orders['Month'])['Revenue'].sum()
print(monthly_sales)
```

## 2.0 Customers Behavior

Similarly, we can sort customers who are producing more revenue

- For that we analyze merge data `merge_sales_orders` according to `CustomerID`

- Here, we use `head()` function to display Top 5 Customers

```python
top_customers = merge_sales_orders.groupby('CustomerID')['Revenue'].sum().sort_values(ascending=False)
print(top_customers.head(5))
```

## 3.0 Product Performance

Similarly, we can sort products that are producing more revenue

- For that we analyze merge data `merge_sales_orders` according to `ProductID`

- Here, we use `head()` function to display Top 5 Products

```python
top_products = merge_sales_orders.groupby('ProductID')['Revenue'].sum().sort_values(ascending=False)
print(top_products.head(5))
```

# C) Feature Engineering

## 1.0 Revenue Per Product

To display Revenue Per Product , we need
- Quantity from orders DataFrame/Table
- Price from products DataFrame/Table

To get data from two DataFrames, we merge both DataFrames

```python
orders.columns
```

```python
products.columns
```

Merge Orders and Product DataFrames

```python
merge_orders_products = pd.merge(orders, products, on='ProductID')
merge_orders_products
```

Revenue Per Product by Quantity * Price

```python
orders['revenue_per_product'] = merge_orders_products['Quantity'] * merge_orders_products['Price']
print(orders.head(5))
```
