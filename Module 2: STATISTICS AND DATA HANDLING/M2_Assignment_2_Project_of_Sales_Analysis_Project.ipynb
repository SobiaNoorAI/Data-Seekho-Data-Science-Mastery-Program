**Project: Sales Analysis Project**

**Objective**

To gain business insights and understand customer and product trends through SQL queries.

**Dataset Description:** A dataset called sales_data containing the following columns:
OrderID (unique identifier for each order)

*   CustomerID (ID of the customer who placed the order)
*   ProductID (ID of the product sold)
*   OrderDate (date of the sale)
*   Quantity (number of products sold in each order)
*   UnitPrice (price per unit of the product)
*   Region (region where the sale was made)
*   Category (category of the product)



```python
---create sales_data table in Xampp
CREATE TABLE sales_data (Order_ID INT(11) PRIMARY KEY AUTO_INCREMENT,
                         CustomerID INT(11),
                         ProductID INT(11),
                         OrderDate DATE,
                         Quantity INT(11),
                         UnitPrice DECIMAL(10,2),
                         Region VARCHAR(255),
                         Category VARCHAR(255)
                        );
```

**Total Sales Revenue**

Find the total revenue generated across all sales.

```python
SELECT SUM(Quantity * UnitPrice) AS TotalRevenue
FROM sales_data;
```

image.png

Here, we find the revenue of all products and stores in the new column `TotalRevenue` .


**Q2: Top 5 Best-Selling Products**

Identify the top 5 products based on the quantity sold.


```python
SELECT ProductID, SUM(Quantity) AS TotalQuantitySold
FROM sales_data
GROUP BY ProductID
ORDER BY TotalQuantitySold DESC
LIMIT 5;
```

image.png

Here, we make groups of `ProductID` and sum their quantities. Then, we place the sum in descending order. Finally, LIMIT 5 displays 5 rows only.

**Q3: Monthly Sales Trend**

Calculate the monthly revenue for each month to identify sales trends.


```python
SELECT
    EXTRACT(YEAR FROM OrderDate) AS Year,
    EXTRACT(MONTH FROM OrderDate) AS Month,
    SUM(Quantity * UnitPrice) AS MonthlyRevenue
FROM sales_data
GROUP BY Year, Month
ORDER BY Year, Month;
```

image.png

In this statement, `YEAR`, and `MONTH` are extracted from `OrderDate` . GROUP BY will display monthly revenue after the new column Year and Month.

**Q4: Regional Sales Breakdown**

Analyze sales revenue for each region to determine where the highest sales are occurring.


```python
SELECT Region, SUM(Quantity * UnitPrice) AS RegionalRevenue
FROM sales_data
GROUP BY Region
ORDER BY RegionalRevenue DESC;
```

image.png

Here `GROUP BY` clause displays the revenue according to `Region` whereas, `ORDER BY` displays revenue in descending order.



**Q5: Most Popular Product Category**

Determine which product category has the highest total sales in revenue.

```python
SELECT Category, SUM(Quantity * UnitPrice) AS CategoryRevenue
FROM sales_data
GROUP BY Category
ORDER BY CategoryRevenue DESC
LIMIT 1;
```

image.png

Here, the `GROUP BY` clause displays the revenue according to `Category` whereas, `ORDER BY` displays revenue in descending order. LIMIT 1 shows only one category.
