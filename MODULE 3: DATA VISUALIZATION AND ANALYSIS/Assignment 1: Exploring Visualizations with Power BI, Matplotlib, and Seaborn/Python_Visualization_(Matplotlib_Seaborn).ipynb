# Matplotlib
It is a Python library to create attractive visualizations Like 2D.

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(dataset['Month'], dataset['Sales'])
plt.show()
```

## 1) Line Plot
This plot is ideal for time-series data

```python
# Line Chart: Revenue Over Months
plt.figure(figsize=(10, 6))
monthly_revenue=data.groupby('Month')['Revenue'].sum()
monthly_revenue.plot(kind='line', marker='o', color='blue')
plt.title('Revenue Over Months', fontsize=16)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Revenue', fontsize=12)
plt.grid()
plt.savefig('monthly_revenue.png')
plt.show()
```

## 2) Bar Plot
This plot is helpful for comparing numeric values.

```python
# Bar Chart: Revenue by Category
plt.figure(figsize=(10, 6))
category_revenue=data.groupby('Category')['Revenue'].sum()
category_revenue.plot(kind='bar', color='orange')
plt.title('Revenue by Category', fontsize=16)
plt.xlabel('Category', fontsize=12)
plt.ylabel('Revenue', fontsize=12)
plt.xticks(rotation=45)
plt.savefig('category_revenue.png')
plt.show()
```

## 3) Histogram
The histogram display the frequency distribution by grouping data into the bins.

```python
# Histogram: Price Distribution
plt.figure(figsize=(10,6))
plt.hist(data['Price'], bins=20, color='skyblue',alpha=1, edgecolor='black')
plt.title('Distribution of Product Prices', fontsize=16)
plt.xlabel('Price', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(axis='y', alpha=0.75)
plt.savefig('price_distribution.png')
plt.show()
```

## 4) Scatter Plot
This plot illustrations the relations between two variables.

```python
# Scatter Plot: Discount By Revenue
plt.figure(figsize=(10,5))
plt.scatter(data['Discounts'], data['Revenue'], color='orange', marker='o')
plt.xlabel('Discounts')
plt.ylabel('Revenue')
plt.title('Discount By Revenue')
#save chart
plt.savefig('discount_revenue.png')
plt.show()
```

```python
# Scatter Plot: Revenue vs Quantity
plt.figure(figsize=(10, 6))
plt.scatter(data['Quantity'], data['Revenue'], alpha=0.7, color='green')
plt.title('Revenue vs Quantity', fontsize=16)
plt.xlabel('Quantity', fontsize=12)
plt.ylabel('Revenue', fontsize=12)
plt.grid()
plt.show()
```

## 5) Pie Chart
The pie plot is a circular graph that display in the form of slices.

```python
# Pie Chart: Revenue by Segment
plt.figure(figsize=(5, 5))
regional_revenue=data.groupby('Region')['Revenue'].sum()
plt.pie(regional_revenue, labels=regional_revenue.index, autopct='%1.1f%%', startangle=140)
plt.title('Revenue Share by Segment', fontsize=16)
plt.savefig('segment_revenue.png')
plt.show()
```

## 6) Box Plot
It displays the minimum, median, and maximum values.

```python
plt.figure(figsize=(10, 6))
# Use 'boxprops' within a dictionary to set the color of the box.
plt.boxplot(data['Revenue'], vert=False, patch_artist=True,  # Enable filled boxes
            boxprops=dict(facecolor='lightblue'))  # Set box color to Red
plt.title('Revenue Distribution', fontsize=16)
plt.xlabel('Revenue', fontsize=12)
plt.grid()
plt.savefig('revenue_boxplot.png')
plt.show()
```

## 7) Area Plot
It is the extension of line plots that shade the area beneath the line.

```python
# Area Chart: Cumulative Revenue Over Months
cumulative_revenue = data.groupby('Month')['Revenue'].sum().cumsum()
plt.figure(figsize=(10, 6))
plt.fill_between(cumulative_revenue.index, cumulative_revenue, color='skyblue', alpha=0.4)
plt.title('Cumulative Revenue Over Months', fontsize=16)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Cumulative Revenue', fontsize=12)
plt.grid()
plt.savefig('cumulative_revenue.png')
plt.show()
```

# Seaborn
Seaborn is an advance version of Matplotlib. it offers the attractive interface for creating statistical visuals.

```python
#import library
import seaborn as sns
```

## 1) Line Plot
Used for time-series data.

```python
# Line Plot
plt.figure(figsize=(10, 6))
sns.lineplot(data=data.groupby('Month')['Revenue'].sum().reset_index(), x='Month', y='Revenue', marker='o')
plt.title('Revenue Trend Over Months')
plt.xlabel('Month')
plt.ylabel('Revenue')
plt.grid(True)
plt.savefig('revenue_trend.png')
plt.show()
```

## 2) Bar Plot
Used for comparing categories.

```python
# Bar Chart
plt.figure(figsize=(10, 6))
sns.barplot(data=data.groupby('Region')['Revenue'].sum().reset_index(), x='Region', y='Revenue' ,hue='Revenue')
plt.title('Revenue by Region')
plt.xlabel('Region')
plt.ylabel('Total Revenue')
plt.savefig('revenue_by_region.png')
plt.show()
```

## 3) Histogram
Displays the data distribution.

```python
# Histogram
plt.figure(figsize=(10, 6))
#kde-->kernel density estimate
sns.histplot(data=data, x='Revenue', bins=20, kde=True, color='green')
plt.title('Revenue Distribution')
plt.xlabel('Revenue')
plt.ylabel('Frequency')
plt.savefig('revenue_distribution.png')
plt.show()
```

## 4) Scatter Plot
Shows the relationships between variables.

```python
# Scatter Plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='Quantity', y='Revenue', hue='Region')
plt.title('Quantity vs. Revenue')
plt.xlabel('Quantity')
plt.ylabel('Revenue')
plt.savefig('quantity_vs_revenue.png')
plt.show()
```

## 5) Pie Plot
There is no default function for the pie chart in seaborn but it provides color\_palette() to make it more attractive.

```python
# Pie Chart
regional_revenue = data.groupby('Region')['Revenue'].sum()
plt.figure(figsize=(5, 5))
plt.pie(regional_revenue, labels=regional_revenue.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('bright'))
plt.title('Revenue Share by Region')
plt.savefig('revenue_share_by_region.png')
plt.show()
```

## 6) Box Plot
Highlights data distribution and outliers.

```python
# Box Plot
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, x='Region', y='Revenue', palette='Set1', legend=True, hue='Region')
plt.title('Revenue Distribution by Region')
plt.xlabel('Region')
plt.ylabel('Revenue')
plt.savefig('revenue_distribution_by_region.png')
plt.show()
```

## 7) Area Plot
Seaborn does not have a built-in function to area charts, we can use Matplotlib’s fill_between() function to create it.

```python
# Area Chart
cumulative_revenue = data.groupby('Month')['Revenue'].sum().cumsum().reset_index()
plt.figure(figsize=(10, 6))
#grid style
sns.set(style="darkgrid")
plt.fill_between(cumulative_revenue['Month'], cumulative_revenue['Revenue'], color='skyblue', alpha=0.4)
#line
plt.plot(cumulative_revenue['Month'], cumulative_revenue['Revenue'], color="Slateblue", alpha=0.7, linewidth=2)
plt.title('Revenue Area Chart Over Months')
plt.xlabel('Month')
plt.ylabel('Revenue')
plt.grid(True)
plt.savefig('revenue_area_chart.png')
plt.show()
```

## 8) Heatmap
The heatmap displays the data magnitude using colors in a grid format. It is useful for identifying patterns and correlations.

```python
# Select only numeric columns for correlation calculation
numerical_data = data.select_dtypes(include=['number']) # This line selects only columns with numeric data types.

# Create a heatmap using Seaborn
plt.figure(figsize=(10, 6))
correlation_matrix = numerical_data.corr() # Calculate correlation on numeric data only.
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linecolor='white', linewidths=1, vmin='0', vmax='1')
plt.title('Correlation Matrix Heatmap')
plt.show()
```

Specific Columns, Monthly Discounts, Revenue, and Taxes Heatmap

```python
# Group data by month and calculate sum of Discounts, Revenue, and Taxes
monthly_data = data.groupby('Month').agg({'Discounts': 'sum', 'Revenue': 'sum', 'Taxes': 'sum'})

# Create the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(monthly_data[['Discounts', 'Revenue', 'Taxes']], annot=True, fmt=".2f", cmap='viridis', linewidths=.5)
plt.title('Monthly Discounts, Revenue, and Taxes Heatmap')
plt.xlabel('Metrics')
plt.ylabel('Month')
plt.savefig('monthly_metrics_heatmap.png')
plt.show()
```
