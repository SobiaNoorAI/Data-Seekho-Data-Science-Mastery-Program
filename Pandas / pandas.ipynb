Import pandas Library

```python
#import pandas library
import pandas as pd
```

Create a Series

```python
# create a list
student_list =['Harry', 'Hermione', 'Ron']

#convert student list into a Series
student_series = pd.Series(student_list)
student_series
```

```python
#access the Harry throgh index
print(student_series[0])
```

Create a DataFrame

```python
#create DataFrame in Key:Value form
student_attendance = pd.DataFrame({"Harry":[1,0,1,1,0], "Hermione":[1,1,1,1,1], "Ron":[1,0,0,0,0]})
student_attendance
```

Change Index

```python
student_attendance = pd.DataFrame({"Harry":[1,0,1,1,0], "Hermione":[1,1,1,1,1], "Ron":[1,0,0,0,0]}
,index=['Monday',' Tuesday',' Wednesday', 'Thursday',' Friday'])
student_attendance
```

Accessing Elements and Slices

```python
#access Monday Label
student_attendance.loc['Monday']
```

```python
#access slice from Monday to Thursday
student_attendance['Monday':'Thursday']
```

```python
#access two columns
student_attendance.columns
```

```python
student_attendance[['Harry', 'Ron']]
```
